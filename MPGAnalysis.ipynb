## Using k-means, Determine how cars are grouped together by using the mpg dataset in Seaborn.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.cluster import KMeans
```

## Load in Data

```python
Mpg = sns.load_dataset('mpg')
```

```python
Mpg.head()
```

## Data Wrangling

```python
mpgTrimmed = Mpg.drop(['origin', 'name'], axis = 1)
```

```python
mpgTrimmed.head()
```

```python
mpgTrimmed.dropna(inplace=True)
```

### I had to come back and drop NA values after a warning when running the k-means

## k-Means Analysis

```python
kmeans = KMeans(n_clusters=2, n_init=10)
kmeans.fit(mpgTrimmed)
```

## Graphing

## First with 2 clusters (high and low)

### MPG & Cylinders

```python
plt.figure(figsize=(10,6))
plt.title('K Means')
plt.scatter(mpgTrimmed['cylinders'], mpgTrimmed['mpg'], c=kmeans.labels_, cmap='cividis')
plt.xlabel('Cylinders')
plt.ylabel('MPG')
```

### MPG & Weight

```python
plt.figure(figsize=(10,6))
plt.title('K Means')
plt.scatter(mpgTrimmed['weight'], mpgTrimmed['mpg'], c=kmeans.labels_, cmap='cividis')
plt.xlabel('Weight')
plt.ylabel('MPG')
```
